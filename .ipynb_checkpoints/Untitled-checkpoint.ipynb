<h2>単回帰分析</h2>

```python
#説明変数と目的変数の指定

exp_var = 'MedInc' # 説明変数
tar_var = 'MedHouseVal' # 目的変数
```

```python
# 散布図を表示
plt.figure(figsize=(12, 9))
plt.scatter(housing_all[exp_var], housing_all[tar_var])
plt.xlabel(exp_var)
plt.ylabel(tar_var)
plt.savefig('california_housing_scatter.png')
```

```python
housing_all[["MedInc","MedHouseVal"]].describe()
```

```python
# 外れ値を除去

q_95 = housing_all['MedInc'].quantile(0.95)
print('95%点の分位数', q_95)

# 絞り込む
data = housing_all[housing_all['MedInc'] < q_95]

sns.boxplot(x=data["MedInc"])

# 散布図を表示
plt.figure(figsize=(12, 9))
plt.scatter(housing_all[exp_var], housing_all[tar_var])
plt.xlabel(exp_var)
plt.ylabel(tar_var)
plt.savefig('california_housing_scatter2.png')
plt.show()
```

```python
# 記述統計量を確認
data[['MedInc', 'MedHouseVal']].describe()
```

```python
# 説明変数と目的変数にデータを分割
X = data[[exp_var]]
print(data.shape)
display(X.head())
y = data[[tar_var]]
print(y.shape)
display(y.head())
```

```python
# 学習
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
# 精度の確認
print('回帰直線の切片', model.intercept_[0])
print('回帰係数', model.coef_[0][0])
print('決定係数', model.score(X, y))
print('回帰直線', 'y = ', model.coef_[0][0], 'x + ', model.intercept_[0])
```

```python
# 回帰直線と散布図を表示
plt.figure(figsize=(12, 9))
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel(exp_var)
plt.ylabel(tar_var)
plt.title('california_housing_regression')
plt.savefig('california_housing_regression.png')
```

<h2>重回帰分析</h2>

```python
# 説明変数
exp_vars = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']
#exp_vars =  ['MedInc', 'HouseAge', 'AveRooms', 'Latitude','Longitude']
# 目的変数
tar_var = "MedHouseVal"
```

```python
#欠陥値の有無確認
housing_all.isnull().sum()
```

```python
# 外れ値を除去
for exp_var in exp_vars:
    q_95 = housing_all[exp_var].quantile(0.95)
    print(exp_var, '95%点の分位数', q_95)
    data = housing_all[housing_all[exp_var] < q_95]
```

```python
# 説明変数と目的変数にデータを分割
X = data[exp_vars]
print(data.shape)
display(X.head())
y = data[[tar_var]]
print(y.shape)
display(y.head())
y = data[[tar_var]]
print(y.shape)
display(y.head())
# 訓練データとテストデータに分割
# 分割のためのモジュールをインポート
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=0)
print('X_train', X_train.shape)
print('y_train', y_train.shape)
print('X_test', len(X_test))
print('y_test', len(y_test))
```

```python
#  X_trainを標準化する
# 標準化ライブラリをインポート
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
print('X_train_scaled', X_train_scaled.shape)
X_train_scaled = pd.DataFrame(X_train_scaled, columns = exp_vars)
display(X_train_scaled.head())
```

```python
# 学習
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# 予測値を計算
y_pred = model.predict(X_train_scaled)
y_pred[:10]

# 偏回帰係数を計算 
for xi, wi in zip(exp_vars, model.coef_[0]):
    print('{0:7s}: {1:6.3f}'.format(xi, wi))
```

```python
# 決定係数
print('決定係数:{:.3f}'.format(model.score(X_train_scaled, y_train)))

from sklearn.metrics import mean_squared_error

# 訓練データに対する Mean Squared Error (MSE)
mse_train = mean_squared_error(y_train, y_pred)
print("訓練データに対するMSE: {:.3f}".format(mse_train))

# テストデータに対する MSE
X_test_scaled = scaler.transform(X_test) # テストデータを訓練データから得られた平均と標準偏差で標準化
y_test_pred = model.predict(X_test_scaled) # テストデータに対して予測する
mse_test = mean_squared_error(y_test, y_test_pred)
print("テストデータに対するMSE: {:.3f}".format(mse_test))
```

```python
## Ridge回帰

from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
ridge_y_pred = ridge.predict(X_train_scaled)
```

```python
# 偏回帰係数の確認
ridge_w = pd.DataFrame(ridge.coef_.T, index=exp_vars, columns=['Ridge'])
for xi, wi in zip(exp_vars, ridge.coef_[0]):
    print('{0:7s}: {1:6.3f}'.format(xi, wi))
print('L2ノルム: {:.3f}'.format(np.linalg.norm(ridge.coef_)))
```

```python
# 訓練データに対する Mean Squared Error (MSE)
ridge_mse_train = mean_squared_error(y_train, ridge_y_pred)
print("訓練データに対するMSE: {:.3f}".format(ridge_mse_train))

# テストデータに対する MSE
ridge_y_test_pred = ridge.predict(X_test_scaled) # テストデータに対して予測する
ridge_mse_test = mean_squared_error(y_test, ridge_y_test_pred)
print("テストデータに対するMSE: {:.3f}".format(ridge_mse_test))

# 決定係数
print('決定係数:{:.3f}'.format(ridge.score(X_train_scaled, y_train)))
```

```python
data = {'訓練データMSE':[mse_train, ridge_mse_train], 
        'テストデータMSE':[mse_test, ridge_mse_test], 
        '決定係数':[model.score(X_test_scaled, y_test), ridge.score(X_test_scaled, y_test)]}
df_mse = pd.DataFrame(data=data, index=['重回帰', 'Ridge回帰'])
df_mse
```

<h2>タイタニック</h2>

```python
#データの読み込み
df = pd.read_csv('./input/titanic3.csv')
```

```python
#ラベルエンコーディング
df_copy =df.copy()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() #ラベルエンコーダのインスタンスを作成

df_copy['sex'] = le.fit_transform(df_copy['sex']) #エンコーディング
#df_copy['embarked'] = le.fit_transform(df_copy['embarked'].astype(str))  #ここ、なぜかstrに変換しないとエラー発生
df_copy['embarked'] = le.fit_transform(df_copy['embarked'])
df_copy.head()

df_copy['embarked']
```

```python
#欠損値確認
df_copy.isnull().sum()
```

```python
#欠損値保管
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].mean()) #欠損値にageの平均値で補完
df_copy['fare'] = df_copy['fare'].fillna(df_copy['fare'].mean()) #欠損値にfareの平均値で補完
print(df_copy.isnull().sum())
```

```python
#不要列の削除
df_copy = df_copy.drop(['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest'],axis=1)
df_copy
```

```python
# ndarray形式への変換

features = df_copy[['pclass','age','sex','fare','embarked']].values
#features = df_copy[['age','sex','fare']].values
target = df_copy['survived'].values
```

```python
# 学習データとテストデータに分割

from sklearn.model_selection import train_test_split

(features , test_X , target , test_y) = train_test_split(features, target , test_size = 0.3 , random_state = 0)
```

```python
#学習の実行

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,random_state=0) # ランダムフォレストのインスタンスを作成

model.fit(features,target) # 学習の実行
```

```python
# 予測の実行
pred = model.predict(test_X)
```

```python
# 【98】予測精度の確認
from sklearn.metrics import accuracy_score

accuracy_score(pred,test_y)
```

```python
# 重要度の確認
# 学習における各列(特徴量)の
# 重要度を表示

importance = model.feature_importances_ 

print('Feature Importances:')
for i, feat in enumerate(['pclass','age','sex','fare','embarked']):
#for i, feat in enumerate(['age','sex','fare']):
    print('\t{0:20s} : {1:>.5f}'.format(feat, importance[i]))
```

```python
df_pred = pd.DataFrame(pred)
df_pred.to_csv('./output/submission.csv',header=None)
```
